````python
def dtw_warp(channels: dict):
    master_channel = list(channels.values())[0] # NOTE: A big assumption is made here: That the first channel passed in here is the "best"
    print(master_channel.counts)
    warped_channel_counts = {}
    for name, chan in channels.items():
        allign = dtw.dtw(x=master_channel.counts, y=chan.counts)
        print(f"Done alligning chan: {name}.")

        warped = chan.counts[allign.index2]
        warped_channel_counts[name] = warped
        print(f"Saved chan: {name} to new")
    return warped_channel_counts
````
